```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/50_startups.csv")
```

```python
dataset
```

```python
# Since the state column has categorical data, need to do the one-hot-encoding
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
X = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
y = dataset["Profit"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# standardization
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.fit(X_test)
```

```python
from sklearn.svm import SVR

regressor = SVR(kernel="linear")
regressor.fit(X_train, y_train)
```

```python
# weights and intercepter
weights = regressor.coef_
bias = regressor.intercept_ # origin
print(weights)
print(bias)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
print(score)
```
