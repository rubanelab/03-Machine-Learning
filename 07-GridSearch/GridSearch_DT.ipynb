### Grid Search using DecisionTree
Inputs and outputs not splited. Directly passing to the model

```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/insurance_pre.csv")
```

```python
dataset
```

```python
dataset.columns
```

### **Pre-Processing**

```python
# The columns sex and smoker has the categorical data needs to do the one-hot-endocing.
# This step will convert the values to numerical
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

### **Separate the input and output variables**

```python
X = dataset[["age", "bmi", "children", "sex_male", "smoker_yes"]]
y = dataset["charges"]
```

```python
X
```

```python
y
```

### DecisionTreeRegressor

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

grid_param = {
                "criterion": ["mse", "mae", "friedman_mse"],
                "max_features": ["auto", "sqrt", "log2"],
                "splitter": ["best", "random"]
             }

grid = GridSearchCV(DecisionTreeRegressor(), grid_param, refit = True, verbose = 3, n_jobs = 1)
grid.fit(X, y)
```

```python
age = float(input("Age: "))
bmi = float(input("BMI: "))
children = float(input("Children: "))
gender = float(input("Gender Male=0, Female=1: "))
smoker = float(input("Smoker Yes=0, No=1: "))

predicted_charges = grid.predict([[age, bmi, children, gender, smoker]])

smoker = "smoking" if smoker == 0 else "non-smoking"
print("Using DecisionTree: {} years old {} {} having {} BMI and {} kids, the charges will be", age, smoker, gender, bmi, children, predicted_charges)
```
