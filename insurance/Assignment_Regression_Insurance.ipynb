```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/insurance_pre.csv")
```

```python
dataset
```

```python
dataset.columns
```

### **Pre-Processing**

```python
# The columns sex and smoker has the categorical data needs to do the one-hot-endocing.
# This step will convert the values to numerical
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

### **Separate the input and output variables**

```python
X = dataset[["age", "bmi", "children", "sex_male", "smoker_yes"]]
y = dataset["charges"]
```

```python
X
```

```python
y
```

### **Split the training and test dataset**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### **1. Multiple Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Multiple Linear Regression:", score)
```

### **2. SVM**

```python
from sklearn.svm import SVR
regressor = SVR(kernel="linear")
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("SVM:", score)
```

### **3. DecisionTree**

```python
from sklearn.tree import DecisionTreeRegressor

# criterion {squared_error, friedman_mse, absolute_error, poisson}
# splitter {best, random}

regressor = DecisionTreeRegressor(criterion='absolute_error', splitter='best')
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Decision Tree (absolute_error, best):", score)
```

```python
regressor = DecisionTreeRegressor(criterion='absolute_error', splitter='random')
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Decision Tree (absolute_error, random):", score)
```

```python
regressor = DecisionTreeRegressor(criterion='friedman_mse', splitter='best')
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Decision Tree (friedman_mse, best):", score)
```

```python
regressor = DecisionTreeRegressor(criterion='friedman_mse', splitter='random')
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Decision Tree (friedman_mse, random):", score)
```

```python
regressor = DecisionTreeRegressor(criterion='absolute_error', splitter='best')
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("Decision Tree (absolute_error, best):", score)
```

### **4. RandomForest**

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

score = r2_score(y_test, y_pred)
print("RandomForest (n_estimators=10):", score)
```
