### **Boosting Algorithms**

```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/insurance_pre.csv")
```

```python
dataset
```

### **Data collection & Pre-processing**

```python
# Encoding: Convert the categorical columns (SEX/SMOKER) into numerical
dataset = pd.get_dummies(dataset, drop_first=True)

# Since the BMI data already within a range as in standard, I not standardize again
```

```python
dataset
```

```python
dataset.columns
```

```python
# Separate the input and output variables
X = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
y = dataset["charges"]
```

```python
X
```

```python
y
```

```python
# Split the training and test dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### 1. Ada Boosting (Adaptive Boosting)

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostRegressor.html

regressor = AdaBoostRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
print("R2 Score for  AdaBoost: ", score)
```

### 2. XGBoost (Extreme Gradient Boosting)

```python
import xgboost as xgb
```

```python
regressor = xgb.XGBRegressor(max_depth=3, learning_rate=0.1, n_estimators=100)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
score = r2_score(y_test, y_pred)
print("XGBoost (Extreme Gradient Boosting): ", score)
```

### 3. LG Boost (Light Gradient Boosting)

```python
# https://www.geeksforgeeks.org/machine-learning/regression-using-lightgbm/

from lightgbm import LGBMRegressor
```

```python
regressor = LGBMRegressor(metric='rmse')
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
score = r2_score(y_test, y_pred)
print("LG Boost (Light Gradient Boosting): ", score)
```
